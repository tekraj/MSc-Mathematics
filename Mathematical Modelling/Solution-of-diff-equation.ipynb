```python
initial_conditions = [1,1.1,0.9,0.001]
solutions = []
for x_1 in initial_conditions:
    x = [x_1]
    for j in range(100):
        x_n = x[j]
        x.append( x_n * (2-x_n)**(1/2) )
    
    solutions.append(x)
    

# plot each solution in its own subplot arranged horizontally
n = len(solutions)
fig, axes = plt.subplots(1, n, figsize=(4*n, 3), sharey=True)
colors = ['blue', 'red', 'green', 'black']

for ax, sol, ic, c in zip(axes, solutions, initial_conditions, colors):
    ax.plot(sol, color=c)
    ax.set_title(f'x0={ic}')
    ax.set_xlabel('iteration')

axes[0].set_ylabel('x')
plt.tight_layout()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def ricker_model(alpha, beta, x0, generations):
    x = np.zeros(generations)
    x[0] = x0
    for n in range(generations - 1):
        x[n+1] = alpha * x[n] * np.exp(-beta * x[n])
    return x

beta = 0.1
generations = 50
x0 = 5

alphas = [1,2.0, 5.0, 8.0] 

plt.figure(figsize=(12, 5))

for i, alpha in enumerate(alphas):
    data = ricker_model(alpha, beta, x0, generations)
    fixed_point = np.log(alpha) / beta if alpha > 1 else 0

    plt.subplot(1, 4, i+1)
    plt.plot(data, '-', linewidth=1.5)
    plt.axhline(y=fixed_point, color='r', linestyle='--', label=f'Fixed Point: {fixed_point:.2f}')
    plt.title(f'Alpha = {alpha}\n(Stability Check)')
    plt.xlabel('Generations (n)')
    plt.ylabel('Population (x)')
    plt.legend()

plt.tight_layout()
plt.show()
```
