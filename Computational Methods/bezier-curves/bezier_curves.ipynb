## Linear Bezier Curve Example

```python
import matplotlib.pyplot as plt
import numpy as np

# Control points

p_0 = (1, 10)
p_1 = (10, 10)



def bezier_linear(t, p0, p1):
    x = (1 - t) * p0[0] + t * p1[0]
    y = (1 - t) * p0[1] + t * p1[1]
    return (x, y)


t_values = np.linspace(0, 1, 100)
bezier_points = [bezier_linear(t, p_0, p_1) for t in t_values]
bezier_x, bezier_y = zip(*bezier_points)


plt.plot(bezier_x, bezier_y, label='Linear Bezier Curve')
plt.plot(*p_0, 'ro', label='P0')
plt.plot(*p_1, 'bo', label='P1')
plt.legend()
plt.title('Linear Bezier Curve')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.axis('equal')  
plt.show()

```

## Quadratic Bezier Curve Example

```python

p_0 = (0, 0)
p_1 = (1, 4)
p_2 = (3, 3)


# draw simple line
def bezier_quad(t, p0, p1, p2):
    # Quadratic Bezier curve formula
    x = (1 - t)**2 * p0[0] + 2 * (1 - t) * t * p1[0] + t**2 * p2[0] # x-coordinate of the Bezier point
    y = (1 - t)**2 * p0[1] + 2 * (1 - t) * t * p1[1] + t**2 * p2[1] # y-coordinate of the Bezier point
    return (x, y)

# Plotting the curve
t_values = np.linspace(0, 1, 100)
bezier_points = [bezier_quad(t, p_0, p_1, p_2) for t in t_values]
bezier_x, bezier_y = zip(*bezier_points)

plt.plot(bezier_x, bezier_y, label='Quadratic Bezier Curve')
plt.plot(*p_0, 'ro', label='P0')
plt.plot(*p_1, 'go', label='P1')
plt.plot(*p_2, 'bo', label='P2')
plt.legend()
plt.title('Quadratic Bezier Curve')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.show()

```

## Cubic Bezier Curve Example

```python
p_0 = (0, 0)
p_1 = (1, 4)
p_2 = (3, -2)
p_3 = (4, 0)


# draw cubic bezier curve
def bezier_cubic(t, p0, p1, p2, p3):
    # Cubic Bezier curve formula
    x = (1 - t)**3 * p0[0] + 3 * (1 - t)**2 * t * p1[0] + 3 * (1 - t) * t**2 * p2[0] + t**3 * p3[0] # x-coordinate of the Bezier point
    y = (1 - t)**3 * p0[1] + 3 * (1 - t)**2 * t * p1[1] + 3 * (1 - t) * t**2 * p2[1] + t**3 * p3[1] # y-coordinate of the Bezier point
    return (x, y)

# Plotting the curve
t_values = np.linspace(0, 1, 100)
bezier_points = [bezier_cubic(t, p_0, p_1, p_2, p_3) for t in t_values]
bezier_x, bezier_y = zip(*bezier_points)

plt.plot(bezier_x, bezier_y, label='Cubic Bezier Curve')
plt.plot(*p_0, 'ro', label='P0')
plt.plot(*p_1, 'go', label='P1')
plt.plot(*p_2, 'bo', label='P2')
plt.plot(*p_3, 'mo', label='P3')
plt.legend()
plt.title('Cubic Bezier Curve')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.show()

```
